```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service as FirefoxService
from selenium.webdriver.firefox.options import Options as FirefoxOptions
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
options = FirefoxOptions()
options.add_argument('--headless')  
service = FirefoxService(executable_path='/opt/homebrew/bin/geckodriver')  
driver = webdriver.Firefox(service=service, options=options)

url = "https://timesofindia.indiatimes.com/topic/sustainable-development/news"
driver.get(url)
def load_more_content():
    while True:
        try:
            load_more_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//button[text()="LOAD MORE ARTICLES"]'))
            )

            driver.execute_script("arguments[0].scrollIntoView(true);", load_more_button)
            load_more_button.click()

            time.sleep(2)
        except Exception as e:
            print(f"Exception: {e}")
            break  
load_more_content()

hrefs = driver.find_elements(By.XPATH, "//div[@class='uwU81 ']/a")

href_links = [link.get_attribute('href') for link in hrefs]

driver.quit()

print(f"Total links found: {len(href_links)}")
print(href_links[:10])  
with open('scraped_links.txt', 'w') as f:
    for href in href_links:
        f.write(href + '\n')
```

```python
art = {"titles":[],"content":[]}
for i in href_links:
    req = requests.get(i)
    res = HtmlResponse("example.com",body = req.text , encoding = 'utf-8')
    con = res.xpath("//div[@class = '_s30J clearfix  ']/text()").getall()
    titl = res.xpath("//h1/span/text()").get()
    con = " ".join(con)
    art['content'].append(con)
    art['titles'].append(titl)
```

```python
import pandas as pd
df = pd.DataFrame(art)
df.to_csv("india.csv")
```

```python
url = ["https://timesofindia.indiatimes.com/topic/no-poverty",
      "https://timesofindia.indiatimes.com/topic/zero-hunger",
      "https://timesofindia.indiatimes.com/topic/good-health-and-well-being",
      "https://timesofindia.indiatimes.com/topic/quality-education",
      "https://timesofindia.indiatimes.com/topic/gender-equality",
      "https://timesofindia.indiatimes.com/topic/clean-water-and-sanitation",
      "https://timesofindia.indiatimes.com/topic/affordable-and-clean-energy",
      "https://timesofindia.indiatimes.com/topic/decent-work-and-economic-growth",
      "https://timesofindia.indiatimes.com/topic/industry,-innovation-and-infrastructure",
      "https://timesofindia.indiatimes.com/topic/reduced-inequalities",
      "https://timesofindia.indiatimes.com/topic/sustainable-cities-and-communities",
      "https://timesofindia.indiatimes.com/topic/responsible-consumption-and-production",
      "https://timesofindia.indiatimes.com/topic/climate-action",
      "https://timesofindia.indiatimes.com/topic/life-below-water",
      "https://timesofindia.indiatimes.com/topic/life-on-land",
      "https://timesofindia.indiatimes.com/topic/peace-justice-and-strong-institutions"]
hrefs = []
for l in url:
    req = requests.get(l)
    res = HtmlResponse("example.com",body = req.text , encoding = "utf-8")
    hrefs.append(res.xpath("//div[@class = 'uwU81 ']/a/@href").getall())
```

```python
hrefs
```

```python
allin = {"title":[],"content":[]}
ti = []
co = []
for i in range(len(hrefs)):
    for l in hrefs[i]:
        r = requests.get(l)
        re = HtmlResponse("example.com",body = r.text , encoding ="utf-8")
        ti.append(re.xpath("//div[@class = 'pZFl7']//h1//text()").getall())
        co.append(re.xpath("//div[@class = '_s30J clearfix  ']/text()").getall())
```

```python
for j in range(len(co)):
    co[j] = ' '.join(co[j])
```

```python
for i in range(len(ti)):
    if ti[i] != []:
        ti[i] = ti[i][0]
    else:
        ti[i] = None
```

```python
for i in range(len(ti)):
    if ti[i] != None:
        allin['title'].append(ti[i])
        allin['content'].append(co[i])
```

```python
allin
```

```python
import pandas as pd
df1 = pd.read_csv("india.csv")
df2 = pd.DataFrame(allin)
```

```python
df2.columns = ['titles','content']
```

```python
main = pd.concat([df1,df2])
```

```python
main.drop("Unnamed: 0" , axis = 1 , inplace = True)
```

```python
main.to_csv("India.csv")
```
