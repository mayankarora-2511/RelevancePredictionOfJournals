```python
msnbc = "https://www.msnbc.com/green"
import requests
from scrapy.http import HtmlResponse
```

```python
r = requests.get(msnbc)
```

```python
res = HtmlResponse("example.com" , body = r.text , encoding = "utf-8")
```

```python
titles = {"title":[]}
urls = []
for j in res.xpath("//span[@class = 'tease-card__headline']/text()"):
    titles['title'].append(j.get())
```

```python
for i in res.xpath("//h2[@class = 'tease-card__headline tease-card__title tease-card__title--msnbc relative']/a/@href"):
    urls.append(i.get())
```

```python
content = {"content":[]}
for u in urls:
    ru = requests.get(u)
    resu = HtmlResponse("example.com" , body = ru.text , encoding = "utf-8")
    for _ in resu.xpath("//span[@class = 'video-details__dek-description']/text()"):
        content['content'].append(_.get())
```

```python
articles = {"titles":titles['title'] , "content":content['content']}
```

```python
detailed_articles = {"title":[],"summary":[],"content":[]}
URL = []
req = requests.get("https://www.nbcnews.com/science/environment")
response = HtmlResponse("example.com" , body = req.text , encoding = "utf-8")
for U in response.xpath("//h2[@class = 'tease-card__headline tease-card__title tease-card__title--news relative']/a/@href"):
    URL.append(U.get())
```

```python
for _ in URL:
    req1 = requests.get(_)
    res1 = HtmlResponse("example.com" , body = req1.text , encoding = "utf-8")
    title = res1.xpath("//h1[@class = 'article-hero-headline__htag lh-none-print black-print']/text()").get()
    summary = res1.xpath("//div[@class = 'summary-box__content']/ul/li/text()").getall()
    article_content = res1.xpath("//div[@class = 'article-body__content'][1]/p/text()").getall()
    detailed_articles['title'].append(title)
    detailed_articles['summary'].append(summary)
    detailed_articles['content'].append(article_content)
```

```python
import pandas as pd 
df1 = pd.DataFrame(articles)
df2 = pd.DataFrame(detailed_articles)
```

```python
df1.to_csv("MSNBC-articles.csv")
df2.to_csv("MSNBC-detailed-articles.csv")
```
